```python
x=dia.drop(columns=['Outcome','SkinThickness','BloodPressure'],axis=1)
x=pd.DataFrame(x)
y=dia['Outcome']
```

```python
x.head()
```

```python
y.head()
```

```python
scale=StandardScaler()
```

```python
x=scale.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30,stratify=y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
svmmodel=svm.SVC(kernel='linear')
```

```python
svmmodel.fit(x_train,y_train)
```

```python
y_pred=svmmodel.predict(x_test)
```

```python
print('Accuracy Score for test data is = ',accuracy_score(y_test,y_pred))
print('confusion matrix =',confusion_matrix(y_test,y_pred))
```

```python
x_pred=svmmodel.predict(x_train)
```

```python
print('Accuracy Score for training data = ',accuracy_score(x_pred,y_train))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100],             
    'kernel': ['linear', 'rbf', 'poly'], 
    'gamma': ['scale', 'auto', 0.01, 0.1] 
}
```

```python
grid = GridSearchCV(svm.SVC(), param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(x_train, y_train)
```

```python
print("best parameter found:", grid.best_params_)
print("best cross val score:", grid.best_score_)
```

```python
best_svm = grid.best_estimator_
test_preds = best_svm.predict(x_test)
print("tuned test accuracy:", accuracy_score(y_test, test_preds))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
models = {
    'Linear SVM': svm.SVC(kernel='linear', random_state=42),
    'RBF SVM': svm.SVC(kernel='rbf', C=1.0, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42),
    'Gradient Boosting': HistGradientBoostingClassifier(random_state=42)
}
```

```python
print("models comparison")
for name, model in models.items():
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    acc = accuracy_score(y_test, predictions)
    print(name ,' model accuracy = ', acc )
```
